## Format results as latex table

```python
latex_table = ''
for country in country_order:
    latex_table += '\multirow{4}{*}{%s} ' % country
    country_df = results[results['country'] == country]
    for metric, err in metric_order:
        latex_table += '& %s ' % metric
        for mapname in map_order:
            map_df = country_df[country_df['dataset'] == mapname]
            rank = country_df[metric].sort_values(ascending=False).unique()
            if map_df[metric].values[0] == rank[0]:
                latex_table += "& \color{{blue}}$\\bm{{{:.2f}\pm{:.2f}}}$ ".format(map_df[metric].values[0], map_df[err].values[0])
            elif map_df[metric].values[0] == rank[1]:
                latex_table += "& $\\bm{{{:.2f}\pm{:.2f}}}$ ".format(map_df[metric].values[0], map_df[err].values[0])
            else:
                latex_table += "& ${:.2f}\pm{:.2f}$ ".format(map_df[metric].values[0], map_df[err].values[0])
        # add the average across all maps for each country
        mean = country_df[country_df['dataset'].isin(map_order)][metric].mean()
        std = country_df[country_df['dataset'].isin(map_order)][err].mean() # Changed from std to mean
        latex_table += "& ${:.2f}\pm{:.2f}$ ".format(mean, std)
        latex_table += '\\\ \n'
    latex_table += '\midrule \n'

# Add the average across all countries for each map
latex_table += '\multirow{4}{*}{Mean}'
for metric, err in metric_order:
    latex_table += '& %s ' % metric
    for mapname in map_order:
        map_df = results[results['dataset'] == mapname]
        mean = map_df[metric].mean()
        std = map_df[err].mean() # Changed from std to mean
        latex_table += '& ${:.2f}\pm{:.2f}$ '.format(mean, std)
    latex_table += '& -- \\\ \n'

latex_table = latex_table.replace('accuracy', 'Accuracy')
latex_table = latex_table.replace('crop_f1', 'F1')
latex_table = latex_table.replace('crop_precision_ua', 'Precision (UA)')
latex_table = latex_table.replace('crop_recall_pa', 'Recall (PA)')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1)

for country, country_df in results.groupby('country'):
    if country == 'Namibia':
        continue
    country_df = country_df[country_df['dataset'].isin(map_order)]
    orig_names = list(country_df.dataset)
    orig_metric = list(country_df.crop_f1)
    orig_err = list(country_df.std_crop_f1)
    new_metric = [orig_metric[j] for j in [orig_names.index(i) for i in map_order]]
    new_err = [orig_err[j] for j in [orig_names.index(i) for i in map_order]]
    ax.errorbar(x=map_order, y=new_metric, yerr=new_err, fmt='o', label=country)

plt.xticks(rotation=90);
plt.legend()
```

### Scatter plot of F1 score vs. spatial resolution

```python
resolution = {'ensemble-subset': 10,
              'esri-lulc': 10,
              'dynamicworld': 10,
              'digital-earth-africa': 10,
              'worldcover': 10,
              'esa-cci-africa': 20,
              'nabil-etal-2021': 30,
              'gfsad-gcep': 30,
              'glad': 30,
              'copernicus': 100,
              'globcover-v23': 300,
              'asap': 1000
             }
```

```python
metric_names = {'accuracy': 'Accuracy',
                'crop_f1': 'F1',
                'crop_precision_ua': 'Precision (User Accuracy)',
                'crop_recall_pa': 'Recall (Producer Accuracy)'}
```

```python
fig, axes = plt.subplots(ncols=4, figsize=(10,3))

axes = axes.flatten()
for i, (metric, err) in enumerate(metric_order):
    res = []
    met = []
    for mapname, mapdf in results.groupby('dataset'):
        if mapname not in map_order:
            continue
        if mapname in ['globcover-v23', 'asap', 'copernicus']:
            continue
        #axes[i].scatter([resolution[mapname]]*mapdf[metric].shape[0], mapdf[metric], label=mapname)
        axes[i].errorbar(resolution[mapname], 
                         mapdf[metric].mean(), 
                         yerr=mapdf[metric].std(), 
                         fmt='o', 
                         label=mapname,
                         color='k',
                         alpha=0.8)
        res.append(resolution[mapname])
        met.append(mapdf[metric].mean())
    axes[i].set_xlabel('Spatial resolution (m/px)')
    axes[i].set_ylabel(metric_names[metric])
    axes[i].set_title('$R^2=${:.2f}'.format(np.corrcoef(res, met)[0,1]))

fig.tight_layout()
```

### Temporal mismatch vs accuracy

```python
year = {
      'esri-lulc': 2019,
      'dynamicworld': 2019,
      'digital-earth-africa': 2019,
      'worldcover': 2020,
      'esa-cci-africa': 2016,
      'nabil-etal-2021': 2016,
      'gfsad-gcep': 2015,
      'glad': 2019,
      'copernicus': 2019,
      'globcover-v23': 2009,
      'asap': 2017
             }
```

```python
fig, axes = plt.subplots(ncols=4, figsize=(10,3))

axes = axes.flatten()
for i, (metric, err) in enumerate(metric_order):
    years = []
    met = []
    for mapname, mapdf in results.groupby('dataset'):
        if mapname not in map_order:
            continue
        if mapname == 'ensemble-subset':
            continue
        for country in mapdf['country']:
            if country == 'Namibia':
                continue
            elif country == 'Malawi':
                mismatch = np.abs(2020-year[mapname])
            else:
                mismatch = np.abs(2019-year[mapname])
            result = mapdf[mapdf['country']==country][metric].values[0]
            axes[i].scatter(mismatch, result, color='k')
            years.append(mismatch)
            met.append(result)
    axes[i].set_xlabel('Temporal mismatch (years)')
    axes[i].set_ylabel(metric_names[metric])
    axes[i].set_title('$R^2=${:.2f}'.format(np.corrcoef(years, met)[0,1]))

fig.tight_layout()
```

## Confusion matrices

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
fig, ax = plt.subplots(nrows=3, ncols=4, figsize=(12,9))
axes = ax.flatten()

for i, mapname in enumerate(map_order):
    total = results[results['dataset'] == mapname][['tn', 'fp', 'fn', 'tp']].sum()
    cm = np.array(total).astype(np.uint32).reshape(2,2)
    p = ConfusionMatrixDisplay(cm).plot(ax=axes[i], cmap='magma_r', im_kw={'vmin':150, 'vmax':3386}, text_kw={'fontsize': 'xx-large'})
    p.im_.colorbar.remove()
    if 'ensemble' in mapname:
        mapname = 'majority-vote'
    axes[i].set_title(mapname, fontsize=14)
    axes[i].set_xlabel('Map label', fontsize=14)
    axes[i].set_ylabel('Reference label', fontsize=14)
    axes[i].tick_params(axis='both', which='major', labelsize=14)
    
fig.suptitle('Confusion Matrices for Combined Reference Dataset', fontsize=20)
fig.tight_layout()
```
